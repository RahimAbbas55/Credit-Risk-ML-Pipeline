```python
import pandas as pd
from src.data_ingestion import load_application_train
pd.set_option('display.max_columns', None)
```

```python
# Loading the data
df = load_application_train()
df.head()
```

```python
print(df.shape)
df['TARGET'].value_counts(normalize=True)
```

```python
# Checking for missing data
missing = df.isnull().sum()
missing_pct = (missing / len(df)) * 100
missing_summary = pd.DataFrame({'missing_count': missing, 'missing_pct': missing_pct})
missing_summary = missing_summary[missing_summary['missing_count'] > 0].sort_values('missing_pct', ascending=False)
missing_summary.head(20)
```

```python
# Looking for key numeric features directly
df[['AMT_INCOME_TOTAL', 'AMT_CREDIT', 'AMT_ANNUITY', 'DAYS_BIRTH', 'DAYS_EMPLOYED']].describe()
```

```python
# Checking the AMT_INCOME_TOTAL Outlier
df['AMT_INCOME_TOTAL'].sort_values(ascending = False).head(10)
```

```python
# Getting the 99th percentile of AMT_INCOME_TOTAL
df['AMT_INCOME_TOTAL'].quantile([0.5 , 0.9 , 0.95 , 0.99 , 1])
```

```python
# Plot for income distribution
import matplotlib.pyplot as plt

plt.figure(figsize = (8 , 5))
df[df['AMT_INCOME_TOTAL'] < 1_000_000]['AMT_INCOME_TOTAL'].hist(bins=50)
plt.title('Distribution of AMT_INCOME_TOTAL (excluding extreme outlier)')
plt.xlabel('Income')
plt.ylabel('Count')
plt.show()
```
